```python
import jieba
import pymongo
from collections import defaultdict
import os
import re
import codecs
client = pymongo.MongoClient("localhost", 27017) # Connect to the db server
db = client.bilibili
db1 = db['danmu']
```

```python
with open(r'C:\Users\Alan\Desktop\2\BosonNLP_sentiment_score.txt','rb') as f:
    sen_list = f.readlines()
    sen_dict = defaultdict()
    for s in sen_list:
        s = s.replace('\n','')
        sen_dict[s.split(' ')[0]] = s.split(' ')[1]
```

```python
senloc = unicode(sen_dict.keys())
for i in db1.find()[:100]:
    score = 0
    print i['content']
    seg = jieba.cut(i['content'])
    seg_result = list()
#     print ','.join(seg)
    for w in seg:
        seg_result.append(w)
    
    with open(r'C:\Users\Alan\Desktop\2\stopwords.txt','rb') as f:
        stopwords = f.readlines()
    stopword = list()
    for k in stopwords:
        a = unicode(k.decode('gbk'))
        a = a.replace('\r\n','')
        stopword.append(a)
        
    new_result = list()
    for word in seg_result:
#         print [word]
        if word in stopword:
            continue
        else:
            new_result.append(word)
    for k in new_result:
        if k in senloc:
            try:
                score += float(sen_dict[k])
            except:
                pass
    print score
```
