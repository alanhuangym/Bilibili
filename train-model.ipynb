```python
import jieba

def readlines(filename):
    fp = open(filename, 'r')
    lines = []
    for line in fp.readlines():
        line = line.strip()
        line = line.decode('utf-8')
        lines.append(line)
    fp.close()
    return lines

def parseSent(sentence):
    # use Jieba to parse sentences
    seg_list = jieba.cut(sentence)
    output = ' '.join(list(seg_list)) # use space to join them
    return output
```

```python
import re  
import codecs  
import os 

# only content is valid
pattern = "<content>(.*?)</content>"
csvfile = codecs.open(r"C:\Users\Alan\Desktop\2\corpus2.csv", 'w', 'utf-8')
fileDir = os.listdir(r"C:\Users\Alan\Desktop\SogouCA")
for file in fileDir:
    with open(r"C:\Users\Alan\Desktop\SogouCA\%s" % file, "r") as txtfile:
        for line in txtfile:
            m = re.match(pattern, line)
            if m:
                segSent = parseSent(m.group(1))
                csvfile.write("%s" % segSent)
```
